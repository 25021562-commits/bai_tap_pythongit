```python
#W4A6
a, b = map(int, input('nhập a,b: ').split())
if a <= 0 or b <= 0:
    print("a và b phải là số nguyên dương")
else:
    if a > b:
        a, b = b, a
    tong = 0
    for n in range(max(2,a), b + 1):
        is_prime = True
        for i in range(2, n):
            if n % i == 0:
                is_prime = False
                break
        if is_prime:
            tong += n
    print(tong)


```

```python
#W4A7
n = int(input('nhập n:'))
if n < 2:
    print('hãy nhập giá trị n >=2:')
else:
    i = 2
    uoc_nguyen_to_lon_nhat = 1
    is_prime = True
    while i <= n:
        if n % i == 0:
            j = 2
            while j <= i // j:  
                if i % j == 0:
                    is_prime = False
                    break
                j += 1
            if is_prime:
                uoc_nguyen_to_lon_nhat = i
        i += 1

    print("Ước số nguyên tố lớn nhất của", n, "là:", uoc_nguyen_to_lon_nhat)

```

```python
#W4A8
n = int(input("Nhập số nguyên dương n: "))

steps = 0

while True:
    temp = n
    rev = 0
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp = temp // 10
    if n == rev:
        break
    n = n + rev
    steps += 1

print("Số bước:", steps)
print("Palindrome thu được:", rev)
```

```python
#4A11
n = int(input("Nhập số nguyên dương n (< 10^6): "))
count = 0
for i in range(2, n + 1, 2):  # chỉ duyệt số chẵn
    if n % i == 0:
        count += 1
print("Số lượng ước số chẵn của", n, "là:", count)
```

```python
#W4A13
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

sum_a = 0
for i in range(1, a):
    if a % i == 0:
        sum_a += i

sum_b = 0
for i in range(1, b):
    if b % i == 0:
        sum_b += i

if sum_a == b and sum_b == a:
    print("true")
else:
    print("false")
```

```python
#W4A14
m = int(input("Nhập số m: "))
n = int(input("Nhập số n: "))
ucln = 1
for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
        ucln = i
print("Ước chung lớn nhất của", m, "và", n, "là:", ucln)
```

```python
#W4A15
total_animals = int(input("Nhập tổng số con: "))
total_legs = int(input("Nhập tổng số chân: "))

found = False
for chickens in range(total_animals + 1):
    dogs = total_animals - chickens
    if chickens * 2 + dogs * 4 == total_legs:
        print("Số gà:", chickens)
        print("Số chó:", dogs)
        found = True
        break

if not found:
    print("invalid")
```
