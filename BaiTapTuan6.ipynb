```python
#W6A9
dict1_input = input("Nhập dict1 ").split()
dict2_input = input("Nhập dict2 ").split()

# Chuyển thành dictionary
dict1 = {}
for pair in dict1_input:
    key, value = pair.split(":")
    dict1[key] = int(value)

dict2 = {}
for pair in dict2_input:
    key, value = pair.split(":")
    dict2[key] = int(value)

# Gộp hai dictionary
dic1copy = dict1.copy()   #tạo bản sao dic1 để dic2 cộng vào không ảnh hưởng dic1
for key in dict2:
    if key in dic1copy:
        dic1copy[key] += dict2[key]
    else:
        dic1copy[key] = dict2[key]

# In ra theo thứ tự từ điển tăng dần
for key in sorted(dic1copy):
    print(f"{key}: {dic1copy[key]}")
```

```python
#W6A10
ds = list(map(int, input("Nhập danh sách số nguyên: ").split()))
k = int(input("Nhập số nguyên dương k: "))


ket_qua = []
n = len(ds)
for i in range(n):
    for j in range(i + 1, n):  # chỉ xét các phần tử sau i
        if ds[i] + ds[j] == k:
            ket_qua.append((ds[i], ds[j]))


ket_qua.sort(key=lambda x: x[0])

print(ket_qua)
```

```python
#W6A12

ds = list(map(int, input("Nhập danh sách số nguyên: ").split()))

dem = {}
for x in ds:
    if x in dem:
        dem[x] += 1
    else:
        dem[x] = 1


max_freq = 0
ket_qua = None
for key in dem:
    if dem[key] > max_freq:
        max_freq = dem[key]
        ket_qua = key
    elif dem[key] == max_freq and key < ket_qua:
        ket_qua = key

print(ket_qua)
```

```python
#W6A13
ds = input("Nhập dictionary (key:value): ").split()


goc = {}
for cap in ds:
    key, value = cap.split(":")
    goc[key] = value


dao = {}
for key in goc:
    dao[goc[key]] = key


print(dao)
```

```python
#W6A15

ds = input("Nhập dictionary (key:value): ").split()
k = int(input("Nhập số nguyên k: "))


goc = {}
for cap in ds:
    key, value = cap.split(":")
    goc[key] = int(value)


loc = {}
for key in goc:
    if goc[key] > k:
        loc[key] = goc[key]

print(loc)
```
