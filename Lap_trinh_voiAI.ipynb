```python
#W1A3
n = 4 
            #Gán giá trị  cho biến . Đây là số cuối của dãy số cần cộng: từ 1 đến 4.
t = 0
            #Khởi tạo biến  với giá trị ban đầu là 0. Biến này sẽ dùng để lưu tổng các số
for i in range(1, n + 1):
    t += i
            #Vòng lặp for chạy từ i=1  đến i=n (tức là 1 đến 4).
            #Mỗi lần lặp, giá trị  được cộng vào biến .
print(t)
            #Sau khi vòng lặp kết thúc, in ra giá trị của t, tức là tổng các số từ 1 đến 4


```

```python
# W1A4
numbers = [1, 2, 3, 4]
                  #Tạo một danh sách gồm 4 số nguyên: 1, 2, 3, 4.
for x in numbers: #Duyệt qua từng phần tử x trong danh sách numbers.
    if x % 2==0:    
                #Kiểm tra xem x có chia hết cho 2 hay không.
                #Nếu đúng (x là số chẵn), thực hiện lệnh trong khối if.
     print(x, "là số chẵn")
    else:
     print(x, "là số lẻ")   
                #Nếu x không chia hết cho 2 (tức là số lẻ), in ra thông báo "là số lẻ".
 
```

```python
# W1A5
animals = ['cat', 'dog', 'cat', 'bird'] 
                 #tạo danh sách gồm 4 phần tử "cat", "dog", "cat", "bird"
count = 0
                 # Khởi tạo biến count với giá trị ban đầu là 0. Biến này sẽ dùng để đếm số phần tử.
for a in animals:
    count += 1
                 # Vòng lặp for duyệt qua từng phần tử a trong danh sách animals.
                 # Mỗi lần lặp, tăng count lên 1 đơn vị.
print("Số phần tử trong danh sách là:", count)
                 # Sau khi vòng lặp kết thúc, in ra tổng số phần tử đã đếm được.
```
