```python
#bài 2
import math
r = float(input("Nhập bán kính hình tròn: "))
s = math.pi * r**2
p = 2 * math.pi * r
print(f"Diện tích hình tròn là: {s:.2f}")
print(f"Chu vi hình tròn là: {p:.2f}")
```

```python
#bài 4  ax+b=0
a = float(input("Nhập hệ số a của phương trình ax+b=0: "))
b = float(input("Nhập hệ số b của phương trình ax+b=0: "))
if a == 0:
    if b == 0:
        print("Phương trình có vô số nghiệm")
    else:
        print("Phương trình vô nghiệm")
if a != 0:
    x = -b / a
    print(f"Nghiệm của phương trình là: x = {x:.2f}")        
```

```python
#bài 5
import math
a = float(input("Nhập hệ số a của phương trình ax^2+bx+c=0: "))
b = float(input("Nhập hệ số b của phương trình ax^2+bx+c=0: "))
c = float(input("Nhập hệ số c của phương trình ax^2+bx+c=0: "))
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c / b
        print(f"Nghiệm của phương trình là: x = {x:.2f}")
if a != 0:
    d = b**2 - 4*a*c
    if d < 0:
        print("Phương trình vô nghiệm")
    elif d == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép: x1 = x2 = {x:.2f}")
    else:
        x1 = (-b + math.sqrt(d)) / (2*a)
        x2 = (-b - math.sqrt(d)) / (2*a)
        print(f"Phương trình có hai nghiệm phân biệt: x1 = {x1:.2f}, x2 = {x2:.2f}")        
```

```python
#bài 6
number = [float(x) for x in input("Nhập 4 số thực, cách nhau bởi dấu cách: ").split()]
max_num = number[0]
for num in number:
    if num > max_num:
        max_num = num
        print("Số lớn nhất trong 4 số đã nhập là:", num)
```

```python
#bài 8
a, b, m = map(int, input("Nhập hệ số a, b ,m của pt ax+by=m, cách bởi dấu cách: ").split())
c, d, n = map(int, input("Nhập hệ số c, d ,n của pt cx+dy=n, cách bởi dấu cách: ").split())
det = a*d - b*c
if det == 0:
    print("Hệ phương trình vô nghiệm hoặc có vô số nghiệm")
else:
    x = (m*d - b*n) / det
    y = (a*n - m*c) / det
    print(f"Nghiệm của hệ phương trình là: x = {x:.2f}, y = {y:.2f}")
```

```python
#bài 9
time = int(input("Nhập số giây: "))
hours = time // 3600
minutes = (time % 3600) // 60   
seconds = time - (hours * 3600 + minutes * 60)
print(f"{hours} giờ, {minutes} phút, {seconds} giây")
```
