```python
#bài 4
h = int(input("Nhập chiều cao tam giác: "))
for i in range(1, h + 1):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")

```

```python
#bài 12
n = int(input("Nhập số nguyên dương n: "))
tong = 0
k = 0

while tong + (k + 1) < n:
    k += 1
    tong += k

print(f"k lớn nhất là {k} với S(k) = {tong}")
```

```python
#bài 14
max_val = min_val = int(input("Nhập số nguyên (-1 để dừng): "))
  #nhập khởi tạo max_val và min_val
while True:
    num = int(input())
    if num == -1:
        break
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Số lớn nhất là:", max_val)
print("Số nhỏ nhất là:", min_val)
```

```python
#bài 18
n = int(input("nhập n:"))
x = 1

while x < n:
    x *= 3

if x == n:
    print(f"{n} là số dạng 3^k")
else:
    print(f"{n} không phải là số dạng 3^k")
```

```python
#bài 19
a = int(input("nhập a:"))
so1 = so2 = 1
while so2 <= a:
    so1, so2 = so2, so1 + so2
print(f"số fibonacci lớn nhất không quá {a} là {so1}")
```

```python
#bài 20
n = int(input("Nhập n: "))
dem = 0

while n > 0:
    cs = n % 10
    if cs in [2, 3, 5, 7]:
        dem += 1
    n //= 10

print("Số chữ số nguyên tố là:", dem) 
```
